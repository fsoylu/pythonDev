```python
# single inhertance

class Hayvanlar:
    def __init__(self, isim):
        self.ad = isim
    
    def hareket_et(self):
        print(f"{self.ad} hareket ediyor")
        
    def beslen(self):
        print(f"{self.ad} besleniyor.")

class Aslanlar(Hayvanlar):
    def __init__(self, isim):
        super().__init__(isim)

    def kukrer(self):
        print(f"{self.ad} kükredi.")
        
    def avlan(self):
        print(f"{self.ad} avlanıyor.")               
       
hayvan = Hayvanlar("Genel Hayvan")
hayvan.hareket_et()
hayvan.beslen()

aslan_kral = Aslanlar("King") 
aslan_kral.kukrer()
aslan_kral.avlan()
aslan_kral.hareket_et()
aslan_kral.beslen()   
```

```python
# multilevel inheritance

class Sekil:
    def __init__(self, renk):
        self.renk = renk

    def alan_hesapla(self):
        print("Alan Hesaplanıyor")

    def ciz(self):
        print(f"{self.renk} renk şekil çiziliyor")

class Dikdortgen(Sekil):
    def __init__(self, renk, kenar_uzunlugu, yukseklik):
        super().__init__(renk)
        self.en = kenar_uzunlugu
        self.boy = yukseklik

    def alan_hesapla(self):
        alan = self.en * self.boy
        print(f"Dikdörtgen alanı: {alan}")

    def ciz(self):
        print(f"{self.renk} renk dikdörtgen çiziliyor")

class Kare(Dikdortgen):
    def __init__(self, renk, kenar_uzunlugu):
        super().__init__(renk, kenar_uzunlugu, kenar_uzunlugu)

    def alan_hesapla(self):
        alan = self.en ** 2
        print(f"Kare alanı: {alan}")

    def ciz(self):
        print(f"{self.renk} renk kare çiziliyor")



sekil = Sekil("Mavi")
sekil.alan_hesapla()
sekil.ciz()

dikdortgen = Dikdortgen("Kırmızı", 20, 50)
dikdortgen.alan_hesapla()
dikdortgen.ciz()

kare = Kare("Mor", 15)
kare.alan_hesapla()
kare.ciz()
```

```python
# Hierarchical Inheritance

class Hayvan:
    def __init__(self, isim):
        self.isim = isim
        
    def kos(self):
        print(f"{self.isim} koştu.")    
    
    def zipla(self):
        print(f"{self.isim} zıpladı.")   
    
    def beslen(self):
        print(f"{self.isim} besleniyor.")
        
class Goriller(Hayvan):
    def __init__(self, isim):
        super().__init__(isim)
        
    def yuz(self):
        print(f"{self.isim} yüzüyor.")  

class Kaplan(Hayvan):
    def __init__(self, isim):
        super().__init__(isim)
        
    def uyu(self):
        print(f"{self.isim} uyuyor.")
        
    def kukre(self):
        print(f"{self.isim} kükrüyor.")
        

hayvan = Hayvan("Genel Hayvan")
hayvan.kos()
hayvan.zipla()
hayvan.beslen()

goril = Goriller("Gaddar")
goril.kos()
goril.zipla()
goril.beslen()
goril.yuz()

kaplan = Kaplan("Adalet")
kaplan.kos()
kaplan.zipla()
kaplan.beslen()
kaplan.kukre()
kaplan.uyu()                 
```

```python
# Multiple Inheritance

class ProglamlamaDili:
    def __init__(self, adi, yaratici, populerlik):
        self.adi = adi
        self.yaratici = yaratici
        self.populerlik = populerlik
        
    def bilgileri_göster(self):
        print("Proglamlama dili Bilgileri:")
        print(f"Adı { self.adi}")
        print(f"Yaratıcısı : {self.yaratici}")
        print(f"Popülerlik : {self.populerlik}")
        
class MuzikEnstrunmani:
    def __init__(self,adi ,turu , calma_sekli):
        self.adi = adi
        self.turu = turu
        self.calma_sekli =calma_sekli
        
    def bilgileri_goster(self):
        print("Müzik Enstürmanı Bilgileri:")
        print(f"Adı : {self.adi}")
        print(f"Türü : {self.turu}")
        print(f"Çalma Şekli : {self.calma_sekli}")
        
class Python(ProglamlamaDili, MuzikEnstrunmani):
    def __init__(self, adi, yaratici, populerlik,turu,calma_sekli,version):
        ProglamlamaDili.__init__(self,adi, yaratici, populerlik)
        MuzikEnstrunmani.__init__(self,adi,turu,calma_sekli)
        self.version = version 
        
    def bilgileri_goster(self):
        print("Python Biilgileri;")
        print(f"Adı         : {self.adi}")
        print(f"Yaratıcısı  : {self.yaratici}")
        print(f"Popülerlik  : {self.populerlik}")
        print(f"Türü        : {self.turu}")
        print(f"Çalma Şekli : {self.calma_sekli}")
        print(f"Version     : {self.version}")
                                               
python = Python("Python","Furkan Soylu", "Yüksek","Genel Amaçlı","Yazılı","3.10.5")
python.bilgileri_goster()    
```

```python
class Personel:

    def __init__(self, ad, soyad, maas):
        self.ad = ad
        self.soyad = soyad
        self.maas = maas
        self.eposta = f"{ad}.{soyad}@firma.com"
        print(f"Yeni Personel tanımlandı: {ad} {soyad}")

    def tam_isim(self):
        return f"{self.ad} {self.soyad}"
    
class Yazilimci(Personel):
    def __init__(self, ad, soyad, maas, pr_dili):
        super().__init__(ad, soyad, maas)
        self.pr_dili = pr_dili
        print(f"Yeni Personel Yazılımcı Kategorisine taşındı {self.tam_isim()}")


class Mudur(Personel):
    def __init__(self, ad, soyad, maas, personeller=None):
        super().__init__(ad, soyad, maas)
        if personeller is None:
            self.personeller = []
        else:
            self.personeller = personeller

    def personel_ekle(self, per):
        if per not in self.personeller:
            self.personeller.append(per)
    
    def personel_cikar(self, per):
        if per in self.personeller:
            self.personeller.remove(per)

    def personelleri_listele(self):
        for per in self.personeller:
            print(per.tam_isim())


yaz_1 = Yazilimci("Berk", "Acar", 40000, "Python")
yaz_2 = Yazilimci("Nihal", "Aktepe", 30000, "Java")

mdr_1 = Mudur("Tolga", "Acar", 50000, [yaz_1])
mdr_2 = Mudur("Koray", "Kanarya", 50000)

print()
print(f"Yeni Müdür: {mdr_1.tam_isim()}")
print("_______________")
print("Personelleri:")
mdr_1.personelleri_listele()
print("_______________")
mdr_1.personel_ekle(yaz_2)
print(f"Yeni personel: {yaz_2.tam_isim()} müdüre tanımlandı")
print("Personelleri:")
mdr_1.personelleri_listele()
mdr_1.personel_cikar(yaz_1)
print(f"{yaz_1.tam_isim()} adlı personel müdürün kadrosundan çıkarıldı")
print("Personelleri:")
mdr_1.personelleri_listele()
```
